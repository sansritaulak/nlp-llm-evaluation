```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load results
df = pd.read_csv('outputs/results/prompt_comparison.csv')
```

```python
# Plot accuracy comparison
plt.figure(figsize=(10, 6))
plt.bar(df['prompt_type'], df['accuracy'])
plt.xlabel('Prompt Type')
plt.ylabel('Accuracy')
plt.title('Prompt Accuracy Comparison')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('outputs/figures/prompt_accuracy.png')
plt.show()
```

```python
# Plot latency vs accuracy
plt.figure(figsize=(10, 6))
plt.scatter(df['avg_latency_sec'], df['accuracy'], s=100)
for i, txt in enumerate(df['prompt_type']):
    plt.annotate(txt, (df['avg_latency_sec'][i], df['accuracy'][i]))
plt.xlabel('Average Latency (seconds)')
plt.ylabel('Accuracy')
plt.title('Latency vs Accuracy Trade-off')
plt.tight_layout()
plt.savefig('outputs/figures/latency_vs_accuracy.png')
plt.show()
```

```python
# Show best prompt
best_idx = df['accuracy'].idxmax()
print(f"\n Best Prompt: {df.loc[best_idx, 'prompt_type']}")
print(f"   Accuracy: {df.loc[best_idx, 'accuracy']:.4f}")
print(f"   Latency: {df.loc[best_idx, 'avg_latency_sec']:.2f}s")
```
